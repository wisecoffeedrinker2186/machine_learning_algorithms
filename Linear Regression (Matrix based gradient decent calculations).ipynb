Linear Regression without regularisation

Adding L2 Regularisation

```python
def predict_y(x,theta):
    return (np.transpose(theta) * x).reshape(-1,1)
```

```python
def get_theta_reg(x,y,theta_initial,alpha,m,Lambda):
    vl = [0]
    vl.extend([1 for i in range((theta.shape[0])-1)])
    tmatrix = np.diag(vl) 
    y_pred = predict_y(x,theta_initial)
    return (theta_initial - (Lambda*np.dot(tmatrix,theta)) - (alpha / m) * np.dot(x ,(y_pred - y)))
```

```python
def error_function(y_pred,y):
    return (np.power((y_pred - y), 2)).sum()
```

```python
def thetaPerIteration(x, y, theta, alpha, m, max_interations,Lambda=0):
    data = {'theta_values': {},
            'j_values': {}}
    for i in range(max_interations):
        y_pred = predict_y(x, theta)
#         theta = get_theta(x, y, theta, alpha, m)
        theta = get_theta_reg(x, y, theta, alpha, m,Lambda)
        error = error_function(y_pred, y)
        if  i != 0 and data['j_values'][i] - error <= 10 ** -3 :
            print(
                f'Tune the learning rate alpha or accept the minimimum value! Interation reached = {i+1},\n', (data['j_values'][i] - error))
            break
        data['theta_values'][i+1] = theta
        data['j_values'][i+1] = error
    return theta, error, data
```

```python
x = np.matrix([[1,1,1,1,1,1],[55,60,65,70,75,80]])
y = np.matrix([[316],[292],[268],[246],[227],[207]])
theta = np.matrix([[400],[-1]])
alpha = 10 ** -3
# alpha = 0.001
m = 6
```

```python
current_theta, current_error, data = thetaPerIteration(x,y,theta,alpha,m,max_interations=100,Lambda=0)
current_theta
```
