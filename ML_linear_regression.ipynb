```python
y=np.matrix(data['Stock_Index_Price']).reshape(-1,1)
x=np.matrix(data.iloc[:,[5,2,3]])
theta=np.matrix(np.random.randint(100, size=(3, 1)))
m=len(x)
alpha=0.001
```

```python
n=int(input("Enter the number of interations:"))
```

```python
def theta_fetch():
  theta=np.matrix(np.random.randint(100, size=(3, 1)))
  return theta
```

```python
def new_theta(theta,x,alpha,m,E):
  theta = theta - np.multiply((alpha/m),(np.transpose(x)*E))
  return theta
```

```python
def y_prediction(theta,x):
  y_pred= x*theta
  return y_pred
```

```python
def Error(E):
  sum=0
  for i in E:
    sum+=(i**2)
  return int(sum)
```

```python
theta=theta_fetch()
j=[]
for i in range(n):
    y_pred=y_prediction(theta,x)
    E=y_pred-y
    theta=new_theta(theta,x,alpha,m,E)
    j.append(Error(E))
```
