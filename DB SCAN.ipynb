```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import PrecisionRecallDisplay,confusion_matrix,f1_score
import pandas as pd
from sklearn.datasets import make_blobs,make_circles,make_moons
from sklearn.cluster import DBSCAN
```

#### DBSCAN Algorithm

```python
X,y = make_blobs(n_samples=250, centers=1, random_state=0, cluster_std=0.3,n_features=2)
```

```python
data = pd.DataFrame({'X1':X[:,0],'X2':X[:,1]})
original_data = data.copy()
data.head(6)
```

```python
plt.scatter(data.X1,data.X2,cmap='winter')
plt.show()
```

```python
dbscan = DBSCAN(eps=0.2,min_samples=20)
```

```python
predicted_values = dbscan.fit_predict(X)
predicted_values
```

```python
data['predicted'] = predicted_values
```

```python
plt.scatter(data.X1,data.X2,c=data.predicted)
plt.show()
```

```python
# dbscan.labels_
```

Noise points have label -1

```python
outlier_value = data[data.predicted == -1]
plt.scatter(data.X1,data.X2,c=data.predicted)
plt.scatter(outlier_value.X1, outlier_value.X2,c='red')
plt.show()
```

If we reduce the eps the more number of outliers are identified making cluseter more accurate

### DBSCAN on Circular Clusters

```python
X,y = make_circles(n_samples=250, random_state=0, noise=0.1, factor=0.2, )
data = pd.DataFrame({'X1':X[:,0],'X2':X[:,1]})
original_data = data.copy()
data.head(6)
```

```python
plt.scatter(data.X1,data.X2)
plt.show()
```

```python
dbscan = DBSCAN(eps=0.2,min_samples=23)
```

```python
predicted_values = dbscan.fit_predict(X)
predicted_values
```

```python
data['predicted'] = predicted_values
```

```python
outlier_value = data[data.predicted == -1]
plt.scatter(data.X1,data.X2,c=data.predicted)
plt.scatter(outlier_value.X1, outlier_value.X2,cmap=data.predicted)
plt.show()
```

```python
for i in range(1,8):
    dbscan = DBSCAN(eps=(i/10),min_samples=25)
    predicted_values = dbscan.fit_predict(X)
    data['predicted'] = predicted_values
    outlier_value = data[data.predicted == -1]
    plt.scatter(data.X1,data.X2,c=data.predicted)
    plt.scatter(outlier_value.X1, outlier_value.X2,cmap=data.predicted)
    plt.show()
```

As we increase the eps all the outliers is considered as normal points and noise detection reduces in accuracy

### DB SCAN on Moon Shaped Data

```python
X,y = make_moons(n_samples=500, random_state=0, noise=0.1)
data = pd.DataFrame({'X1':X[:,0],'X2':X[:,1]})
original_data = data.copy()
data.head(6)
```

```python
plt.scatter(data.X1,data.X2,c=y,cmap='winter')
plt.show()
```

```python
dbscan = DBSCAN(eps=0.19,min_samples=20)
predicted_values = dbscan.fit_predict(X)
# predicted_values
data['predicted'] = predicted_values
outlier_value = data[data.predicted == -1]
plt.scatter(data.X1,data.X2,c=data.predicted)
plt.scatter(outlier_value.X1, outlier_value.X2,cmap=outlier_value.predicted,c='red')
plt.show()
```

### Testing different values for eps and sample size

```python
dbscan = DBSCAN(eps=0.2,min_samples=30)
predicted_values = dbscan.fit_predict(X)
# predicted_values
data['predicted'] = predicted_values
outlier_value = data[data.predicted == -1]
plt.scatter(data.X1,data.X2,c=data.predicted)
plt.scatter(outlier_value.X1, outlier_value.X2,cmap=outlier_value.predicted,c='red')
plt.show()
```

We can infer that as we increase the eps more points are considered to be normal, as sample size decreases the noise and densities vary creating more clusters. To get 2 clusters we keel eps 0.2 - 0.3 and the sample size from 20-24

### DB SCAN on Diabetes dataset

```python
diabetes_data = pd.read_csv('diabetes.csv')
diabetes_data.head()
```

```python
diabetes_data.columns
```

```python
X = diabetes_data[['Glucose', 'BMI']]
y = diabetes_data['Outcome']
X.head()
```

```python
diabetes_data.info()
```

```python
diabetes_data.describe()
```

```python
plt.scatter(X.Glucose,X.BMI,cmap='winter',c=y)
plt.show()
```

```python
dbscan = DBSCAN(eps=5,min_samples=5)
predicted_values = dbscan.fit_predict(X)
# np.where(predicted_values == -1)
predicted_values
```

```python
X['predicted'] = predicted_values
outlier_value = X[X.predicted == -1]
plt.scatter(X.Glucose,X.BMI,c=X.predicted)
plt.scatter(outlier_value.Glucose, outlier_value.BMI,cmap=X.predicted,c='red')
plt.show()
```

In this case the noise is identified correctly when eps is more than sample size
